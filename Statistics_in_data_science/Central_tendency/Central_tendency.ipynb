```python
mn=np.mean(df["age"])
md=df["age"].median()
mo=df["age"].mode()
print("mode is :",mo)
print("median is :",md)
print("mean is :",mn)
sns.histplot(x="age",data=df, bins=[i for i in range(1,80,10)])
plt.plot([mn for i in range(0,500)],[i for i in range(0,500)],c="red",label="mean")
plt.plot([md for i in range (0,500)],[i for i in range(0,500)],c="purple",label="median")
plt.plot([mo for i in range(0,300)],[i for i in range (0,300)],c="green",label="mode")
plt.show()
```

#Mean Absolute Deviation using observation not by calculation

```python

sec_a=np.array([75,65,73,68,72,76])
sec_b=np.array([90,47,43,96,93,51])
no=np.array([1,2,3,4,5,6])
mean=np.mean(sec_a)
print("mean of both is same:",mean)
```

```python
plt.scatter(sec_a,no,label="Sec A",color="red")
plt.scatter(sec_b,no,label="Sec B",color="green")
plt.plot([70,70,70,70,70,70],no,label="Mean_line",color="black")
plt.legend()
plt.show() 
#therefore we select Class A due less spreadness
```

#Mean Absolute Deviation using observation  by calculation

```python

mad_a=np.sum(np.abs(sec_a-mean))/len(sec_a)
mad_b=np.sum(np.abs(sec_b-mean))/len(sec_b)
```

#hence we have conclude that smaller the mean abs deviation higher the chances 

```python
mad_a,mad_b   
```

#std and varience low value indicates that tends to close to mean 

```python
np.std(sec_a),np.std(sec_b)
```

```python
np.var(sec_a),np.var(sec_b)
```

#correlation(-1,1) And Covarience (minus infinity,plus infinity)

```python
data_cov=df.select_dtypes(["float64","int64"]).cov() # values lies between minus infinity to plus infinity 
data_cov
```

```python
data_corr=df.select_dtypes(["float64","int64"]).corr()  #values lies between 0 to 1 
data_cov
```

#to show corr() and cov() we use heatmap

```python
sns.heatmap(data_cov,linecolor="white")
plt.show()
```

```python
sns.heatmap(data_corr)
plt.show()
```
